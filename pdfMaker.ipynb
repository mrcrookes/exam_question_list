```python
# Import libaries
from pypdf import PdfReader, PdfWriter

# Read in raw pdfs
qpaper = PdfReader("./pdfs/N5_Physics_all_2014.pdf")
mpaper = PdfReader("./pdfs/mi_N5_Physics_all_2014.pdf")

# List of question pages and marking instruction pages
qpageslist = [(3, 12), (18, 19), (20, 22)]
mpageslist = [(7, 7), (8, 10), (11, 12)]

# Check if lists of questions and answers have the same length
if len(qpageslist) == len(mpageslist):
    print('List lengths match!')
    
# Create file for pdf
fileA = PdfWriter()

# Loop through list of question pages
for q in range(len(qpageslist)):
    
    # Add all pages for question q
    for n in range(qpageslist[q][0] - 1, qpageslist[q][1]):
        fileA.add_page(qpaper.pages[n])
    
    # Add all pages for question q marking instructions
    for n in range(mpageslist[q][0] - 1, mpageslist[q][1]):
        fileA.add_page(mpaper.pages[n])

with open("./output/test.pdf", "wb") as out:
    fileA.write(out)
```
