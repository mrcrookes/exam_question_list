```python
# Import librarys
import pandas as pd
import numpy as np
```

```python
# Read in csv
df = pd.read_csv('./data/test.csv')

df.head()
```

```python
# Create new df with only Section B questions
dfab = df[(df['Section'] == 'b') | (df['Section'] == 'a')].copy()

# Fill NaN values with empty strings
dfab = dfab.fillna('')

# Convert numbers into integers
dfab = dfab.astype({'Question': int, 'Marks': int, 'Page': int})

# Join 3 questions columns together into 1 new column
dfab["FullQuestion"] = dfab[['Question', 'Letter', 'Part']].agg(lambda x: ''.join(str(v) for v in x), axis=1)

dfab.head()
```

```python
# Group questions together by topic
gp = dfab.groupby(['Subject', 'Year', 'Section']).aggregate({'FullQuestion':lambda x: ', '.join(str(v) for v in x)})

# Unstack headings
gp = gp.unstack()

# Fill NaN values with empty strings
gp = gp.fillna('')

gp.head()
```

```python
gp.to_csv('./csvs/output.csv')
```
