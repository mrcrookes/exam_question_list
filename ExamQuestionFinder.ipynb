```python
# Import librarys
import pandas as pd
import numpy as np
```

```python
# Read in csv
df = pd.read_csv('./data/test.csv')
df.head()
```

```python
# Create new df with only Section B questions
dfb = df[df['Section'] == 'b'].copy()

# Fill NaN values with empty string
dfb = dfb.fillna('')

# Join 3 questions columns together into 1 new column
dfb["FullQuestion"] = dfb[['Question', 'Letter', 'Part']].agg(lambda x: ''.join(str(v) for v in x), axis=1)
```

```python
# Group questions together by topic
gp = dfb.groupby(['Subject', 'Year']).aggregate({'FullQuestion':lambda x: ','.join(str(v) for v in x)})
gp.head()
```

```python
gp.to_csv('./csvs/output.csv')
```
